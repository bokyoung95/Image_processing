```python
from def_bk import *
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('test4.jpg', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('test5.png', cv2.IMREAD_GRAYSCALE)
img2[img2>0]=1
```

## Binary

```python
b = binary_bk(img,128)
```

```python
plt.figure(figsize=(8,16))
plt.subplot(121) #1행2열 중 첫 번째
plt.imshow(img, cmap = 'gray')
plt.subplot(122)#1행2열 중 두 번째
plt.imshow(b, cmap ='gray')
```

## Binary Morphology

### Dilation

```python
s = np.array([[0,0,0],[1,1,1],[0,1,1]])
dil = BmorphD_bk(img2,s)
```

```python
plt.figure(figsize=(8,16))
plt.subplot(121) #1행2열 중 첫 번째
plt.imshow(img2, cmap = 'gray')
plt.subplot(122)#1행2열 중 두 번째
plt.imshow(dil, cmap ='gray')
```

### Erosion

```python
s = np.array([[0,0,0],[1,1,1],[0,1,1]])
count= sum(sum(s))
er = BmorphE_bk(img2,s,count)
```

```python
plt.figure(figsize=(8,16))
plt.subplot(121) #1행2열 중 첫 번째
plt.imshow(img2, cmap = 'gray')
plt.subplot(122)#1행2열 중 두 번째
plt.imshow(er, cmap ='gray')
```

### Opening

```python
s = np.array([[0,0,0],[1,1,1],[0,1,1]])
count= sum(sum(s))
opening = BmorphD_bk(BmorphE_bk(img2,s,count),s)
```

```python
plt.figure(figsize=(8,16))
plt.subplot(121) #1행2열 중 첫 번째
plt.imshow(img2, cmap = 'gray')
plt.subplot(122)#1행2열 중 두 번째
plt.imshow(opening, cmap ='gray')
```

### Closing

```python
s = np.array([[0,0,0],[1,1,1],[0,1,1]])
count= sum(sum(s))
closing = BmorphE_bk(BmorphD_bk(img2,s),s,count)
```

```python
plt.figure(figsize=(8,16))
plt.subplot(121) #1행2열 중 첫 번째
plt.imshow(img2, cmap = 'gray')
plt.subplot(122)#1행2열 중 두 번째
plt.imshow(closing, cmap ='gray')
```
