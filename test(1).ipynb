```python
from def_bk import *
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

## Filter_Smoothing

### Box Filter

```python
box = BoxF_bk(3)
box_c = convolution_bk(img,box)
```

```python
plt.figure(figsize=(8,16))
plt.subplot(121) #1행2열 중 첫 번째
plt.imshow(img_data, cmap = 'gray')
plt.subplot(122)#1행2열 중 두 번째
plt.imshow(box_c, cmap ='gray')
```

### Weighted Mean Filter

```python
wm = WeightedMean_bk()
wm_c = convolution_bk(img,wm)

plt.figure(figsize=(8,16))
plt.subplot(121) 
plt.imshow(img_data, cmap = 'gray')
plt.subplot(122)
plt.imshow(wm_c, cmap ='gray')
```
