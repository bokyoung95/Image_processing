```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('test2.png', cv2.IMREAD_GRAYSCALE)
height = img.shape[0]
width = img.shape[1]
img_data = np.array(img) # 이미지 pixel 데이터를  array로
Hx = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
Hy = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
img_x = np.zeros((height-2,width-2))
img_y = np.zeros((height-2,width-2))
```

```python
for i in range(1,height-1):
    for j in range(1,width - 1):
        sum_s = 0
        for l in range(-1,2):  #필터 사이즈 -> 해당 이미지 픽셀에서 -1부터 1까지 (3by3)
            for k in range(-1,2):
                a = img_data[i+l][j+k] * Hx[l+1][k+1] 
                sum_s = sum_s + a  
        img_x[i-1][j-1] = sum_s
```

```python
plt.imshow(img_x, cmap = 'gray')
```

```python
for i in range(1,height-1):
    for j in range(1,width - 1):
        sum = 0
        for l in range(-1,2):  #필터 사이즈 -> 해당 이미지 픽셀에서 -1부터 1까지 (3by3)
            for k in range(-1,2):
                a = img_data[i+l][j+k] * Hy[l+1][k+1] 
                sum = sum + a   
        img_y[i-1][j-1] = sum
```

```python
plt.imshow(img_y, cmap = 'gray')
```
