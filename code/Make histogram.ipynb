```python
import numpy as np
from PIL import Image
import os
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def hist_bk(Filename):
    img = cv2.imread(Filename, cv2.IMREAD_GRAYSCALE)
    height = img.shape[0]
    width = img.shape[1]
    img_data = np.array(img) # 이미지 pixel 데이터를  array로
    hist_data = np.zeros(256)
    x_g = np.arange(256)
    
    for x in range(height):
        for y in range(width):
            value = img_data[x][y]
            hist_data[value] = hist_data[value] + 1
    

    plt.figure()
    plt.bar(x_g,hist_data, align = 'center', width = 0.3, color = 'b')
    plt.xlabel('pixel')
    plt.ylabel('num')
    plt.xlim(0,len(hist_data))
    plt.ylim(0,hist_data.max())
    plt.show()
    return img_data, hist_data
```

```python
Filename = 'test4.jpg'
```

```python
bk = hist_bk(Filename)
```

```python
plt.imshow(bk[0], cmap = 'gray')
```
