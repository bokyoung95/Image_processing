```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def Gausian_bk(size,sigma):
    Gaussian_filter = np.zeros([size,size])
    for i in range(1,-2, -1):
        for j in range(-1,2):
            Gaussian_filter[i+1,j+1] = np.exp(-(i**2 + j**2)/(2*sigma**2))
    return Gaussian_filter
```

```python
img = cv2.imread('test4.jpg', cv2.IMREAD_GRAYSCALE)
height = img.shape[0]
width = img.shape[1]
img_data = np.array(img) # 이미지 pixel 데이터를  array로
img_g = np.zeros((height-2,width-2))
for i in range(1,height-1):
    for j in range(1,width - 1):
        sum_g = 0
        for l in range(-1,2):  #필터 사이즈 -> 해당 이미지 픽셀에서 -1부터 1까지 (3by3)
            for k in range(-1,2):
                a = img_data[i][j] * Gaussian_filter[l+1][k+1] 
                sum_g = sum_g + a       
        img_g[i-1][j-1] = sum_g
```

```python
plt.imshow(img_data, cmap = 'gray')
```
