```python
import numpy as np
from PIL import Image
import os
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('test4.jpg', cv2.IMREAD_GRAYSCALE)
height = img.shape[0]
width = img.shape[1]
img_data = np.array(img) # 이미지 pixel 데이터를  array로
hist_data = np.zeros(256)
x_g = np.arange(256)
    
for x in range(height):
    for y in range(width):
        value = img_data[x][y]
        hist_data[value] = hist_data[value] + 1
hist_e = np.zeros(256)
hist_round = np.zeros(256)
```

```python
PDF = hist_data /(height * width)
```

```python
for i in range(256):
    CDF = 0
    for j in range(i+1):
        CDF = CDF + PDF[j]
    hist_round[i] = round(255*CDF)    
    
```

```python
img_data
```

```python
for x in range(height):
    for y in range(width):
        value = img_data[x][y]
        img_data[x][y] = hist_round[value]
```

```python
hist_rdata = np.zeros(256)
for x in range(height):
    for y in range(width):
        value = img_data[x][y]
        hist_rdata[value] = hist_rdata[value] + 1
```
