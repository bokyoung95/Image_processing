```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('test2.png', cv2.IMREAD_GRAYSCALE)
height = img.shape[0]
width = img.shape[1]
img_data = np.array(img) # 이미지 pixel 데이터를  array로
img_median = np.zeros((height-2,width-2))
```

```python
for i in range(1,height-1):
    for j in range(1,width - 1):
        temp = []
        for l in range(-1,2):  #필터 사이즈 -> 해당 이미지 픽셀에서 -1부터 1까지 (3by3)
            for k in range(-1,2):
                a = img_data[i+l][j+k]
                temp.append(a) 
        temp.sort()
        value_m = temp[4]
        img_median[i-1][j-1] = value_m
```

```python
plt.imshow(img, cmap = 'gray')
print(img_data)
```

```python
plt.imshow(img_median, cmap = 'gray')
print(img_median)
```
