```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('test3.png', cv2.IMREAD_GRAYSCALE)
```

```python
height = img.shape[0]
width = img.shape[1]
img_data = np.array(img) # 이미지 pixel 데이터를  array로
weighted_filter = np.zeros((height-2,width-2))
```

```python
weight = np.array([[1,2,1],[2,4,2],[1,2,1]])
weight = weight /16
```

```python
for i in range(1,height-1):
    for j in range(1,width - 1):
        sum_w = 0
        for l in range(-1,2):  #필터 사이즈 -> 해당 이미지 픽셀에서 -1부터 1까지 (3by3)
            for k in range(-1,2):
                a = img_data[i+l][j+k] * weight[l+1][k+1] #원본이미지를 기준으로 -1,0,1만큼의 이동 픽셀 값 (height, width)
                sum_w = sum_w + a       
        b = sum_w / 16.0
        weighted_filter[i-1][j-1] = b
```

```python
plt.imshow(img_data, cmap = 'gray')
```

```python
plt.imshow(weighted_filter, cmap = 'gray')
```
